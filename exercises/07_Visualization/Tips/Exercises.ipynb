### Step 1. Import the necessary libraries:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from urllib.request import urlretrieve
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

```python
urlretrieve(
    'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv', '../../../data/tips.csv')
```

### Step 3. Assign it to a variable called tips

```python
tips_df = pd.read_csv('../../../data/tips.csv')
```

### Step 4. Delete the Unnamed 0 column

```python
del tips_df['Unnamed: 0']
tips_df.head()
```

### Step 5. Plot the total_bill column histogram

```python
total_bills = tips_df[['total_bill']]
plt.figure(figsize=(9,5))
sns.set_style('darkgrid')
sns.histplot(total_bills, bins=10);
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.figure(figsize=(9, 5))
sns.scatterplot(x=tips_df.total_bill, y=tips_df.tip, hue=tips_df.sex);
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
# plt.figure(figsize=(9, 5))
sns.pairplot(tips_df)
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.figure(figsize=(9, 5))
sns.barplot(x=tips_df.day, y=tips_df.total_bill)
plt.title('Total Bills Per Day')
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
plt.figure(figsize=(9, 5))
sns.stripplot(data=tips_df, x='tip', y='day', hue='sex')
plt.title('Tips Per Day By Gender');
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
plt.figure(figsize=(9, 5))
sns.boxplot(x=tips_df.day, y=tips_df.total_bill, hue=tips_df.time)
plt.title('Total Bills Per Meal Time');
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
lunch = tips_df[['tip']][tips_df.time == 'Lunch']
dinner = tips_df[['tip']][tips_df.time == 'Dinner']
fig, axs = plt.subplots(1, 2, figsize=(12,6), sharex=True, sharey=True)
ax1 = axs[0]
ax2 = axs[1]
sns.histplot(data=lunch, ax=ax1, bins=10)
ax1.set(xlabel='Lunch')
sns.histplot(data=dinner, ax=ax2, bins=10)
ax2.set(xlabel='Dinner')
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
male = tips_df[['tip', 'total_bill', 'smoker']][tips_df.sex == 'Male']
female = tips_df[['tip', 'total_bill', 'smoker']][tips_df.sex == 'Female']
fig, axs = plt.subplots(1, 2, figsize=(12, 6), sharex=True, sharey=True)
ax1 = axs[0]
ax2 = axs[1]
sns.scatterplot(x=male.total_bill, y=male.tip, hue=male.smoker, ax=ax1)
ax1.set_title('Tips to Total Bills for Males')
sns.scatterplot(x=female.total_bill, y=female.tip, hue=female.smoker, ax=ax2)
ax2.set_title('Tips to Total Bills for Females')
```
